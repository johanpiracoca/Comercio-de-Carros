## 2. Analisis de Datos

### 2.1 Distribuciones

```python
# Cantidad de autos clasificados por color y tipo
total_color = car_data.groupby('paint_color')['type'].value_counts()
df_total_color = total_color.reset_index(name='car_count')
df_total_color
```

```python
# Grafico de carros sugun su tipo, segmentados por colores
bar_plot = px.bar(
    df_total_color,
    x = 'type',
    y = 'car_count',
    labels = {
        'type': 'Tipo de automóvil',
        'car_count': 'Cantidad',
        'paint_color': 'Colores'
        },
    title = "Clasificacion de autos segun su tipo y color",
    color = 'paint_color',
    color_discrete_map={
        'black': 'black',
        'blue': 'blue', 
        'brown': 'brown',
        'custom': 'gray',  
        'green': 'green',
        'grey': 'gray',
        'orange': 'orange',
        'purple': 'purple',
        'red': 'red',
        'silver': 'silver',
        'white': 'lightgray', 
        'yellow': 'yellow'
    },
    width = 700,
    height = 400
)

bar_plot.show()
```

```python
# Estandarizacion de la variable Odometer para graficar el histograma 
df_odometer = car_data.dropna(subset=['odometer'])
print(f"Valores nulos: {df_odometer['odometer'].isnull().sum()}")
```

```python
# Histograma de la variable odometer clasificado por el tipo de combustible
hist_plot = px.histogram(
    df_odometer, 
    x = 'odometer', 
    labels = {'fuel': 'Tipo de Combustible'},
    title = f"Distribución de Kilometraje clasificado Tipo de Combustible", 
    color = 'fuel',
    color_discrete_map={
        'diesel': 'black',
        'hybrid': 'blue', 
        'electric': 'green',
        'gas': 'red',
        'other': 'lightgray'
    }, 
    marginal = "box", 
    width = 700, 
    height = 400
)
hist_plot.show()
```

```python
# Calculo de Estadisticos de la variable Odometer
prom_odo = np.mean(df_odometer['odometer'])
med_odo = np.median(df_odometer['odometer'])
desv_odo = np.std(df_odometer['odometer'])
print("Media: {:.1f}".format(prom_odo))
print("Mediana: {:.1f}".format(med_odo))
print("Desviación: {:.1f}".format(desv_odo))
```

### 2.2 Relacion de Variables

```python
# Grafico de dispersion
disp_plot = px.scatter(
    car_data, 
    x = 'model_year', 
    y = 'price', 
    labels = {'model_year': 'Año de Modelo',
              'price': 'Precio',
              'transmission': 'Transmision'},
    color = 'transmission', 
    title = f"Dispersión Año de Modelo vs. Precio", 
    color_discrete_map={  
        'automatic': 'green',
        'manual': 'orange',
        'other': 'lightgray'},
    width = 700, 
    height = 400
)
disp_plot.show()
```

```python
# Calcular correlacion entre variables
correlacion = car_data['model_year'].corr(car_data['price'])
print(f"Correlación de Pearson: {correlacion*100:.2f} %")
```
